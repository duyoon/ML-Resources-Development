```python
from math import sqrt
from numpy import concatenate
from matplotlib import pyplot
from pandas import read_csv
from pandas import DataFrame
from pandas import concat
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.callbacks import ModelCheckpoint
import numpy as np
import tensorflow as tf
import os
```

```python
# 랜덤 시드 조정
random_seed = 10

np.random.seed(random_seed)
tf.set_random_seed(random_seed)
```

```python
# 시계열자료를 학습자료로 변환하기
def convert_data_to_sequence(data, n_in=1, n_out=1, dropnan=True):
    n_vars = 1 if type(data) is list else data.shape[1]
    df1 = DataFrame(data[:,0])
    df2 = DataFrame(data[:,1])
    cols, names = list(), list()
    
    # 학습자료 중 입력 시퀀스 (t-n_in, ..., t-1)
    for i in range(n_in, 0, -1):
        cols.append(df1.shift(i))
        cols.append(df2.shift(i-1))
        names += [('var%d(t-%d)' % (j+1, i)) for j in range(n_vars)]
        
    # 학습자료 중 출력 시퀀스 (t, t+1, ..., t+n_out-1)
    for i in range(0, n_out):
        cols.append(df1.shift(-i))
        cols.append(df2.shift(-i+1))
        if i == 0:
            names += [('var%d(t)' % (j+1)) for j in range(n_vars)]
        else:
            names += [('var%d(t+%d)' % (j+1, i)) for j in range(n_vars)]
            
    # 하나로 합치기
    agg = concat(cols, axis=1)
    agg.columns = names
    
    # NaN 값이 있는 행 삭제
    if dropnan:
        agg.dropna(inplace=True)
    return agg
```

```python
# 시계열 자료 불러오기
dataset = read_csv('../dataset/BHP_ratedata.v3.csv', header=0, index_col=0)
print(dataset.head())
values = dataset.values
values = values.astype('float32')
```

```python
# 학습자료의 표준화
scaler = MinMaxScaler(feature_range=(0, 1))
scaled = scaler.fit_transform(values)
print(scaled[0:5])
```

```python
# 입력자료 당 시간간격 수
n_time = 12
# 각 시간별 특성값의 수
n_features = 2
# 시계열 자료를 입력자료의 형태로 변환
reframed = convert_data_to_sequence(scaled, n_time, 1)
values = reframed.values

print(reframed.head())
```

```python
# 학습자료, 검증자료, 시험자료로 분리
n_total = values.shape[0]
n_valid = 15
n_test = 15
n_train = n_total - (n_valid+n_test)
train = values[:n_train, :]
validation = values[n_train:(n_train+n_valid), :]
test = values[-n_test:, :]

# 입력자료와 목표값으로 분리
n_obs = n_time * n_features
train_X, train_y = train[:, :n_obs], train[:, -n_features]
validation_X, validation_y = validation[:, :n_obs], validation[:, -n_features]
test_X, test_y = test[:, :n_obs], test[:, -n_features]

# 입력자료 형태 변경 -> [samples, n_time, features]
train_X = train_X.reshape((train_X.shape[0], n_time, n_features))
validation_X = validation_X.reshape((validation_X.shape[0], n_time, n_features))
test_X = test_X.reshape((test_X.shape[0], n_time, n_features))

print(train_X.shape, train_y.shape)
print(validation_X.shape, validation_y.shape)
print(test_X.shape, test_y.shape)
```

```python
# 모델 저장 폴더 설정
MODEL_DIR = './model/'
if not os.path.exists(MODEL_DIR):
   os.mkdir(MODEL_DIR)

modelpath="./model/bestWeights_BHP_1.h5"

# 모델을 파일로 저장할지 점검함
checkpointer = ModelCheckpoint(filepath=modelpath, monitor='val_loss', 
                               verbose=1, save_best_only=True) 
```

```python
# 모델 구축
model = Sequential()
model.add(LSTM(100, input_shape=(n_time, n_features)))
model.add(Dense(1))
model.compile(loss='mae', optimizer='adam')

# 모델 학습
history = model.fit(train_X, train_y, epochs=1000, batch_size=6, 
                    validation_data=(validation_X, validation_y), 
                    callbacks=[checkpointer], verbose=1, shuffle=False)

# 최고의 모델을 파일에서 복원
model.load_weights(modelpath)

```

```python
# 학습이력 추적
pyplot.figure()
pyplot.plot(history.history['loss'], label='train')
pyplot.plot(history.history['val_loss'], label='validation')
pyplot.xlabel('epoch')
pyplot.ylabel('error')
pyplot.legend()
pyplot.show()

# 최종모델의 오차 계산
tr_err = model.evaluate(train_X,train_y)
val_err = model.evaluate(validation_X,validation_y)
test_err = model.evaluate(test_X,test_y)
print('train error = %.5f'%tr_err)
print('validation error = %.5f'%val_err)
print('test error = %.5f'%test_err)
```

```python
# 표준화된 자료의 역변환 함수 정의
def inv_scaler (x, y):
    inv_y = concatenate((y, x.reshape(x.shape[0],n_time*n_features)[:, -1:]), 
                        axis=1)
    inv_y = scaler.inverse_transform(inv_y)
    return inv_y[:,0]
        
# 시험자료의 역변환
test_ypred = model.predict(test_X)
inv_test_ypred = inv_scaler(test_X, test_ypred)
inv_test_y = inv_scaler(test_X, test_y.reshape((len(test_y), 1)))
# 검증자료의 역변환
validation_ypred = model.predict(validation_X)
inv_validation_ypred = inv_scaler(validation_X, validation_ypred)
inv_validation_y = inv_scaler(validation_X, 
                              validation_y.reshape((len(validation_y), 1)))
# 학습자료의 역변환
train_ypred = model.predict(train_X)
inv_train_ypred = inv_scaler(train_X, train_ypred)
inv_train_y = inv_scaler(train_X, train_y.reshape((len(train_y), 1)))

# 공저압력의 목표값과 모델의 예측값에 대해 학습자료, 검증자료, 시험자료를 합치기
lead_y = dataset.values[0:n_time,0].reshape(n_time)  #
true_y = np.concatenate([lead_y,inv_train_y,inv_validation_y, inv_test_y], 
                        axis = 0)
pred_y =  np.concatenate([inv_train_ypred,inv_validation_ypred
                          , inv_test_ypred], axis = 0)
```

```python
# 공저압력 목표값과 모델 값의 비교 그래프 그리기
def plot_cmp_results (true_y, pred_y):
    pyplot.figure(figsize=(10,6))
    tnum = len(true_y)
    pyplot.scatter(range(tnum),true_y,color='blue',marker='o',
                   facecolors='none',label='true')
    pyplot.scatter(range(len(lead_y),tnum),pred_y,color='red',label='model')

    yrange = pyplot.ylim()
    tx = len(lead_y) + n_total - n_test 
    pyplot.plot([tx,tx],yrange,'-.')
    pyplot.annotate('test data',xy=(tx,yrange[1]))

    tx = tx -n_valid
    pyplot.plot([tx,tx],yrange,'-.')
    pyplot.annotate('valid data',xy=(tx,yrange[1]))

    tx = len(lead_y)
    pyplot.plot([tx,tx],yrange,'-.')
    pyplot.annotate('train data',xy=(tx,yrange[1]))

    pyplot.legend(loc=4)
    pyplot.xlabel('Time (month)')
    pyplot.ylabel('BHP (psi)')
    pyplot.show()

plot_cmp_results (true_y, pred_y)
```

```python
# 검증자료와 시험자료를 합침
pred_X = np.concatenate ([validation_X, test_X], axis = 0)
Z = pred_X[:1]  # 본래 데이터의 첫번째 자료만 사용함

predicted = []
length_of_test = pred_X.shape[0]

# 검증자료와 시험자료에 대해 모델예측 적용
for i in range(length_of_test):
    z_ = Z[-1:]
    y_ = model.predict(z_)
    # 모델에서 나온 값을 입력자료에 사용
    if i < (length_of_test-1):
        pred1 = np.concatenate([y_,pred_X[i+1,-1:,-1:]],axis=1)
        sequence_ = np.concatenate(
            (z_.reshape(n_time, n_features)[1:], pred1),
            axis=0).reshape(1, n_time, n_features)
        Z = np.append(Z, sequence_, axis=0)
    predicted.append(y_.reshape(-1))

# 표준화 역변환
pred_X1 = pred_X.reshape((pred_X.shape[0], n_time*n_features))
inv_predicted_y = inv_scaler(pred_X1, predicted)
inv_val_test_y = np.concatenate ([inv_validation_y, inv_test_y],axis = 0)
```

```python
# 공저압력 목표값과 모델 값의 비교 그래프 그리기
plot_cmp_results (true_y, 
                  np.concatenate ([inv_train_ypred,inv_predicted_y],axis=0))
```
